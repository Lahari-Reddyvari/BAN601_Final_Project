```python
import pandas as pd
import numpy as np
import seaborn as sns
import streamlit as st
import matplotlib.pyplot as plt
```

```python
#Title
st.title("Sales Performance Dashboard for Accessories")
#Loading dataset
df = pd.read_csv("sales_data.csv")
#View dataset
df.head()
st.subheader("Dataset")
st.write(df)
```

```python
print("Hello")

#
year = st.selectbox(
    "Select Year",
    df["Year"].unique()
)

filtered_df = df[df["Year"] == year]

st.write(filtered_df)

#product category

product = st.selectbox(
    "Select Product",
    ["Handbags_Sales", "Wallets_Sales"]
)

import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots()

sns.lineplot(
    data=filtered_df,
    x="Month",
    y=product,
    marker="o",
    ax=ax
)

ax.set_title(f"{product} Sales Trend")

st.pyplot(fig)


total_sales = filtered_df[product].sum()

st.metric("Total Sales", total_sales)
```
